Создаем функцию анализа данных. 

Проведение комплексного анализа датасета, включая:

- автоматическое преобразование столбцов с датами в datetime64
    
- общую информацию о данных
    
- анализ числовых признаков с визуализацией
    
- проверку на выбросы с учетом отрицательных значений
    
- анализ пропущенных значений
    
</div>

```python
def analyze_dataset(data):
    """
    Проводит комплексный анализ датасета, включая:
    - автоматическое преобразование столбцов с датами в datetime64
    - общую информацию о данных
    - анализ числовых признаков с визуализацией
    - проверку на выбросы с учетом отрицательных значений
    - анализ пропущенных значений
    """
    try:
        # 1. Автоматическое преобразование столбцов с датами
        date_columns = []
        for col in data.select_dtypes(include=['object']).columns:
            try:
                # Пробуем преобразовать столбец в datetime
                data[col] = pd.to_datetime(data[col])
                date_columns.append(col)
                display(f"Столбец '{col}' успешно преобразован в datetime64")
            except (ValueError, TypeError):
                continue

        if date_columns:
            display("\nПреобразованные столбцы с датами:")
            date_info = pd.DataFrame({
                'Столбец': date_columns,
                'Тип данных': [str(data[col].dtype) for col in date_columns],
                'Минимальная дата': [data[col].min() for col in date_columns],
                'Максимальная дата': [data[col].max() for col in date_columns],
                'Пропущено значений': [data[col].isnull().sum() for col in date_columns]
            })
            display(date_info)
        else:
            display("\nСтолбцы с датами не обнаружены или не требуют преобразования")

        # 2. Вывод первых 5 строк данных
        display("\nПервые 5 строк данных:")
        display(data.head())

        # 3. Вывод последних 5 строк данных
        display("\nПоследние 5 строк данных:")
        display(data.tail())

        # 4. Вывод основной информации о данных
        display("\nИнформация о данных:")
        display(data.info())

        # 5. Вывод количества уникальных значений для каждого признака (кроме 'key')
        non_key_columns = [col for col in data.columns if col != 'key']
        display("\nКоличество уникальных значений по столбцам (исключая 'key'):")
        unique_counts = data[non_key_columns].nunique().to_frame(name='Уникальных значений')
        display(unique_counts)

        # 6. Анализ числовых признаков (исключая 'key')
        display("\nАнализ числовых признаков:")
        numeric_cols = [col for col in data.select_dtypes(include=['number']).columns if col != 'key']

        if len(numeric_cols) > 0:
            # Основные статистики с добавлением 99-го перцентиля
            display("\nОсновные статистики числовых признаков:")
            stats = data[numeric_cols].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]).transpose()
            stats['skew'] = data[numeric_cols].skew()
            stats['kurtosis'] = data[numeric_cols].kurtosis()
            display(stats)

            # Проверка на пропуски
            display("\nКоличество пропущенных значений в числовых признаках:")
            missing_values = data[numeric_cols].isnull().sum().to_frame(name='Пропусков')
            display(missing_values[missing_values['Пропусков'] > 0])

            # Анализ распределения
            display("\nАнализ распределения числовых признаков:")
            distribution_info = pd.DataFrame({
                'Медиана': data[numeric_cols].median(),
                'Мода': data[numeric_cols].mode().iloc[0],
                'Асимметрия': data[numeric_cols].skew(),
                'Эксцесс': data[numeric_cols].kurtosis(),
                'Диапазон': data[numeric_cols].max() - data[numeric_cols].min(),
                '99-й перцентиль': data[numeric_cols].quantile(0.99)
            })
            display(distribution_info)

            # Визуализация числовых признаков
            display("\nВизуализация числовых признаков:")
            for col in numeric_cols:
                plt.figure(figsize=(15, 5))
                
                # Гистограмма с ядром плотности
                plt.subplot(1, 2, 1)
                sns.histplot(data[col], kde=True, bins=30)
                plt.title(f'Гистограмма {col}')
                plt.axvline(data[col].mean(), color='r', linestyle='--', label=f'Среднее: {data[col].mean():.2f}')
                plt.axvline(data[col].median(), color='g', linestyle='-', label=f'Медиана: {data[col].median():.2f}')
                plt.axvline(data[col].quantile(0.99), color='b', linestyle=':', label=f'99%: {data[col].quantile(0.99):.2f}')
                plt.legend()
                
                # Боксплот
                plt.subplot(1, 2, 2)
                sns.boxplot(x=data[col])
                plt.title(f'Боксплот {col}')
                
                plt.tight_layout()
                plt.show()
                display("---")

            # Проверка на выбросы
            display("\nАнализ выбросов (метод IQR):")
            for col in numeric_cols:
                Q1 = data[col].quantile(0.25)
                Q3 = data[col].quantile(0.75)
                IQR = Q3 - Q1
                original_lower_bound = Q1 - 1.5 * IQR
                lower_bound = max(original_lower_bound, 0)
                upper_bound = Q3 + 1.5 * IQR

                negative_outliers = data[data[col] < original_lower_bound]
                final_outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]

                display(f"Признак: {col}")
                display(f"Границы выбросов: [{lower_bound:.4f}, {upper_bound:.4f}]")
                display(f"99-й перцентиль: {data[col].quantile(0.99):.4f}")

                if len(negative_outliers) > 0:
                    display(f"Отрицательные выбросы: {len(negative_outliers)}")
                    display("Примеры:", negative_outliers[[col]].head(3))

                display(f"Всего выбросов: {len(final_outliers)} ({len(final_outliers)/len(data)*100:.2f}%)")
                if len(final_outliers) > 0:
                    display("Примеры:", final_outliers[[col]].head(3))
                display("---")
        else:
            display("В данных нет числовых признаков для анализа.")

        # 7. Анализ пропущенных значений
        display("\nАнализ пропущенных значений (исключая 'key'):")
        missing_info = pd.DataFrame({
            'Пропущено значений': data[non_key_columns].isnull().sum(),
            'Процент пропусков': (data[non_key_columns].isnull().mean() * 100).round(2)
        }).sort_values('Пропущено значений', ascending=False)

        display(missing_info[missing_info['Пропущено значений'] > 0])

    except Exception as e:
        display(f"Произошла ошибка при анализе данных: {e}")

```

```python
# Столбец с ключом
key_column = 'key'

# Вычисляем время нагрева в минутах
data_arc_new['Время нагрева (мин)'] = (
    data_arc_new['Конец нагрева дугой'] - data_arc_new['Начало нагрева дугой']
).dt.total_seconds() / 60

# Простая энергия = Активная мощность * Время нагрева (в часах)
data_arc_new['Энергия нагрева (МВт*ч)'] = (
    data_arc_new['Активная мощность'] * (data_arc_new['Время нагрева (мин)'] / 60)
)

# Суммарная энергия (корень из суммы квадратов P и Q) * время в часах
data_arc_new['Полная энергия (МВт*ч)'] = (
    np.sqrt(data_arc_new['Активная мощность']**2 + data_arc_new['Реактивная мощность']**2)
    * (data_arc_new['Время нагрева (мин)'] / 60)
)

# Отношение реактивной мощности к активной (Q/P)
data_arc_new['Отношение Q/P'] = (
    data_arc_new['Реактивная мощность'] / data_arc_new['Активная мощность']
)

# Замена бесконечных значений (при P=0) на NaN
data_arc_new['Отношение Q/P'] = data_arc_new['Отношение Q/P'].replace([np.inf, -np.inf], np.nan)

# Добавляем новые признаки времени
data_arc_new['Час начала плавки'] = data_arc_new['Начало нагрева дугой'].dt.hour
data_arc_new['День недели начала плавки'] = data_arc_new['Начало нагрева дугой'].dt.dayofweek  # 0-понедельник, 6-воскресенье
data_arc_new['День недели начала плавки (название)'] = data_arc_new['Начало нагрева дугой'].dt.day_name()

# Показываем первые 5 строк с новыми колонками (включая ключ)
print("Данные с новыми признаками:")
display_cols = [key_column, 'Активная мощность', 'Реактивная мощность', 'Время нагрева (мин)',
               'Энергия нагрева (МВт*ч)', 'Полная энергия (МВт*ч)', 'Отношение Q/P',
               'Час начала плавки', 'День недели начала плавки', 'День недели начала плавки (название)']
display(data_arc_new[display_cols].head(10))
display(data_arc_new[display_cols].tail(10))

# Описательная статистика новых признаков (группировка по ключу при необходимости)
print("\nОписательная статистика новых признаков:")
display(data_arc_new[['Энергия нагрева (МВт*ч)', 'Полная энергия (МВт*ч)', 'Отношение Q/P',
                     'Час начала плавки', 'День недели начала плавки']].describe())

# Дополнительная проверка - группировка по ключу (если нужно)
print("\nПример агрегации по ключу (сумма энергий):")
display(data_arc_new.groupby(key_column)[['Энергия нагрева (МВт*ч)', 'Полная энергия (МВт*ч)']].sum().head())

# Анализ новых временных признаков
print("\nРаспределение записей по часам:")
display(data_arc_new['Час начала плавки'].value_counts().sort_index())

print("\nРаспределение записей по дням недели:")
display(data_arc_new['День недели начала плавки (название)'].value_counts())
```

Данные о времени загрузки не позволяют определить период самого процесса. Есть только временные метки. Поэтому создаем два новых признака 'Bulk_before', 'Bulk_after'. Это интервал соответственно от начала плавки (первого включения нагрева) до первой загрузки шихтовых материалов и интервал от последней загрузки до последнего включения нагрева.

Заметна зависимость - последняя загрузка из Bulk обычно совпадает с последним включением нагрева.

</div>


```python
# Копируем исходные данные без изменений
data_bulk_time = data_bulk_time_new.copy()

# Выбираем только временные столбцы Bulk
bulk_columns = [col for col in data_bulk_time.columns if 'Bulk' in col and col != 'key']

# Преобразуем временные признаки в data_arc_new в datetime
data_arc_times = data_arc_new.copy()
data_arc_times['Начало нагрева дугой'] = pd.to_datetime(data_arc_times['Начало нагрева дугой'])
data_arc_times['Конец нагрева дугой'] = pd.to_datetime(data_arc_times['Конец нагрева дугой'])

# Получаем минимальные и максимальные времена нагрева
arc_times = data_arc_times.groupby('key').agg({
    'Начало нагрева дугой': 'min',
    'Конец нагрева дугой': 'max'
}).reset_index()

# Объединяем с нашими данными
data_bulk_time = pd.merge(data_bulk_time, arc_times, on='key', how='left')

# Преобразуем Bulk-столбцы в datetime
for col in bulk_columns:
    data_bulk_time[col] = pd.to_datetime(data_bulk_time[col])

# Функция для вычисления интервалов
def calculate_intervals(row):
    # Находим минимальное и максимальное Bulk-время (исключая NaT)
    bulk_times = row[bulk_columns]
    valid_times = bulk_times[bulk_times.notna()]

    if len(valid_times) == 0:
        return pd.Series([np.nan, np.nan], index=['Bulk_before', 'Bulk_after'])

    min_bulk = valid_times.min()
    max_bulk = valid_times.max()

    # Bulk_before: интервал между началом нагрева и первым Bulk
    before = np.nan
    if pd.notna(row['Начало нагрева дугой']):
        before = (min_bulk - row['Начало нагрева дугой']).total_seconds() / 60

    # Bulk_after: интервал между последним Bulk и концом нагрева
    after = np.nan
    if pd.notna(row['Конец нагрева дугой']):
        after = (row['Конец нагрева дугой'] - max_bulk).total_seconds() / 60

    return pd.Series([before, after], index=['Bulk_before', 'Bulk_after'])

# Применяем функцию для создания новых признаков
intervals = data_bulk_time.apply(calculate_intervals, axis=1)
data_bulk_time = pd.concat([data_bulk_time, intervals], axis=1)

# Удаляем временные колонки из data_arc_new
data_bulk_time = data_bulk_time.drop(['Начало нагрева дугой', 'Конец нагрева дугой'], axis=1)

# Проверка на повторяющиеся номера партий (key)
duplicates = data_bulk_time[data_bulk_time.duplicated('key', keep=False)]

# --- Вывод результатов ---
print("Первые 5 строк после обработки:")
display(data_bulk_time.head())

print("\nПоследние 5 строк после обработки:")
display(data_bulk_time.tail())

print("\nСтатистика по новым признакам:")
display(data_bulk_time[['Bulk_before', 'Bulk_after']].describe())

print("\nКоличество пропусков:")
print(f"Bulk_before: {data_bulk_time['Bulk_before'].isna().sum()}")
print(f"Bulk_after: {data_bulk_time['Bulk_after'].isna().sum()}")

print("\nПовторяющиеся номера партий (key):")
if not duplicates.empty:
    display(duplicates[['key'] + bulk_columns + ['Bulk_before', 'Bulk_after']])
else:
    print("Повторяющихся партий не найдено.")

print("\nТипы данных:")
display(data_bulk_time.dtypes)
```

```python
# Удаление партий с пропусками температуры и проблемных партий
def clean_data(data):
    """Функция очистки данных"""
    # 1. Находим партии с пропусками температуры
    temp_missing_keys = data[data['Температура'].isna()]['key'].unique()

    # 2. Список проблемных партий для удаления
    bad_parties = [867, 1214, 1619, 2052]

    # 3. Объединяем все ключи для удаления
    all_bad_keys = list(temp_missing_keys) + bad_parties

    # 4. Удаляем все проблемные партии
    clean_data = data[~data['key'].isin(all_bad_keys)].copy()

    # 5. Выводим статистику
    display(f"Удалено партий с пропусками температуры: {len(temp_missing_keys)}")
    display(f"Удалено указанных проблемных партий: {len(bad_parties)}")
    display(f"Всего удалено строк: {len(data) - len(clean_data)}")
    display(f"Осталось строк: {len(clean_data)}")

    return clean_data

# Применяем очистку данных
data_temp = clean_data(data_temp_new)

# Информация о очищенном датасете
display("\nОчищенный датасет:")
display(data_temp.info())

# Первые строки очищенных данных
display("\nПервые 5 строк очищенных данных:")
display(data_temp.head(10))
```

```python
# Копируем исходные данные
data_wire_proc = data_wire_time.copy()

# Выбираем только временные столбцы Wire
wire_columns = [col for col in data_wire_proc.columns if 'Wire' in col and col != 'key']

# Преобразуем временные признаки в data_arc_new в datetime
data_arc_times = data_arc_new.copy()
data_arc_times['Начало нагрева дугой'] = pd.to_datetime(data_arc_times['Начало нагрева дугой'])
data_arc_times['Конец нагрева дугой'] = pd.to_datetime(data_arc_times['Конец нагрева дугой'])

# Получаем временные границы нагрева
arc_times = data_arc_times.groupby('key').agg({
    'Начало нагрева дугой': 'min',
    'Конец нагрева дугой': 'max'
}).reset_index()

# Объединяем данные
data_wire_proc = pd.merge(data_wire_proc, arc_times, on='key', how='left')

# Преобразуем Wire-столбцы в datetime
for col in wire_columns:
    data_wire_proc[col] = pd.to_datetime(data_wire_proc[col])

# Функция для расчета интервалов
def calc_wire_intervals(row):
    # Фильтруем валидные времена подачи проволоки
    wire_times = row[wire_columns]
    valid_times = wire_times[wire_times.notna()]

    if len(valid_times) == 0:
        return pd.Series([np.nan, np.nan], index=['Wire_before', 'Wire_after'])

    min_wire = valid_times.min()
    max_wire = valid_times.max()

    # Wire_before: от начала нагрева до первой подачи проволоки
    before = np.nan
    if pd.notna(row['Начало нагрева дугой']):
        before = (min_wire - row['Начало нагрева дугой']).total_seconds() / 60

    # Wire_after: от последней подачи проволоки до конца нагрева
    after = np.nan
    if pd.notna(row['Конец нагрева дугой']):
        after = (row['Конец нагрева дугой'] - max_wire).total_seconds() / 60

    return pd.Series([before, after], index=['Wire_before', 'Wire_after'])

# Создаем новые признаки
wire_intervals = data_wire_proc.apply(calc_wire_intervals, axis=1)
data_wire_proc = pd.concat([data_wire_proc, wire_intervals], axis=1)

# Удаляем временные колонки из data_arc_new
data_wire_proc = data_wire_proc.drop(['Начало нагрева дугой', 'Конец нагрева дугой'], axis=1)

# Вычисляем медианные значения (только для информации, без заполнения)
median_before = data_wire_proc['Wire_before'].median()
median_after = data_wire_proc['Wire_after'].median()

# --- Анализ результатов ---
print("Первые 5 строк после обработки:")
display(data_wire_proc.head())

print("\nПоследние 5 строк после обработки:")
display(data_wire_proc.tail())

print("\nСтатистика новых признаков:")
stats = data_wire_proc[['Wire_before', 'Wire_after']].describe()
display(stats)

print(f"\nМедианные значения (только для справки):")
print(f"Wire_before: {median_before:.2f} мин")
print(f"Wire_after: {median_after:.2f} мин")

print("\nРаспределение интервалов:")
print(f"Отрицательные Wire_before: {(data_wire_proc['Wire_before'] < 0).sum()}")
print(f"Отрицательные Wire_after: {(data_wire_proc['Wire_after'] < 0).sum()}")

print("\nПовторяющиеся ключи:")
if not duplicates.empty:
    display(duplicates[['key'] + wire_columns + ['Wire_before', 'Wire_after']])
else:
    print("Дубликатов не найдено")

print("\nТипы данных:")
display(data_wire_proc.dtypes)
```

```python
def prepare_final_dataframe(df):
    """
    Подготавливает финальный датафрейм с очисткой данных, удаляя строки с пропусками
    
    Параметры:
    df - исходный датафрейм
    
    Возвращает:
    Очищенный датафрейм df_final
    """
    # Сохраняем копию исходного датафрейма
    df_final = df.copy()
    
    # Удаляем строки с любыми пропусками
    initial_rows = len(df_final)
    df_final = df_final.dropna()
    removed_rows = initial_rows - len(df_final)
    
    # Выводим информацию о процессе обработки
    print(f"Исходное количество строк: {initial_rows}")
    print(f"Количество строк после очистки: {len(df_final)}")
    print(f"Удалено строк: {removed_rows}")
    print(f"Удалено столбцов: 0")  # Столбцы не удаляются
    
    # Анализ пропусков после обработки
    missing_after = df_final.isnull().sum()
    if missing_after.sum() == 0:
        print("Пропусков в данных не обнаружено")
    else:
        print("Столбцы с пропусками:")
        display(missing_after[missing_after > 0])
    
    return df_final

# Применяем функцию
df_final = prepare_final_dataframe(df)

# Дополнительный анализ данных
print("\nТипы данных и общая информация:")
display(df_final.info())

print("\nОписательная статистика числовых признаков:")
display(df_final.describe(include='number').T)

print("\nПримеры данных (первые 5 строк):")
display(df_final.head(5))

# Проверяем наличие температурных столбцов перед выводом
temp_cols = ['Конечная температура', 'Начальная температура']
existing_temp_cols = [col for col in temp_cols if col in df_final.columns]

if existing_temp_cols:
    print("\nРаспределение температурных признаков:")
    display(df_final[existing_temp_cols].describe())
else:
    print("\nТемпературные признаки не найдены в данных")
```

```python
# 1. Создаем пайплайн с предобработкой и моделью
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Масштабирование признаков
    ('rf', RandomForestRegressor(random_state=140725, n_jobs=-1))  # Модель Random Forest
])

# 2. Определяем сетку гиперпараметров для подбора
param_grid = {
    'rf__n_estimators': [100, 200],  # Количество деревьев
    'rf__max_depth': [None, 10, 20],  # Максимальная глубина
    'rf__min_samples_split': [2, 5],  # Минимальное количество образцов для разделения
    'rf__min_samples_leaf': [1, 2],  # Минимальное количество образцов в листе
    'rf__max_features': ['sqrt', 0.5]  # Количество признаков для поиска лучшего разделения
}

# 3. Настраиваем кросс-валидацию с минимизацией MAE
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring=mae_scorer,  # Минимизируем MAE
    cv=5,  # 5-кратная кросс-валидация
    n_jobs=-1,  # Используем все ядра процессора
    verbose=1  # Вывод информации о процессе
)

# 4. Обучаем модель с подбором параметров
print("Начинаем подбор параметров Random Forest...")
grid_search.fit(X_train, y_train)
print("Подбор параметров завершен!")

# 5. Выводим лучшие параметры
print("\nЛучшие параметры модели:")
best_params = grid_search.best_params_
for param, value in best_params.items():
    print(f"{param}: {value}")

# 6. Оцениваем модель на тренировочных данных
best_model = grid_search.best_estimator_
y_train_pred = best_model.predict(X_train)
train_mae = mean_absolute_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)  # Добавлена метрика R2

print("\nМетрики на тренировочной выборке:")
print(f"MAE: {train_mae:.4f}")
print(f"R2: {train_r2:.4f}")  # Вывод R2

# 7. Кросс-валидация лучшей модели
print("\nРезультаты кросс-валидации лучшей модели:")

# MAE (остаётся как есть)
cv_scores_mae = cross_val_score(
    best_model,
    X_train,
    y_train,
    scoring=mae_scorer,
    cv=5,
    n_jobs=-1
)
cv_mae_scores = -cv_scores_mae

# Добавляем R2 в кросс-валидацию
r2_scorer = make_scorer(r2_score)
cv_scores_r2 = cross_val_score(
    best_model,
    X_train,
    y_train,
    scoring=r2_scorer,
    cv=5,
    n_jobs=-1
)

# Вывод результатов
print(f"Средний MAE: {cv_mae_scores.mean():.4f} (±{cv_mae_scores.std():.4f})")
print(f"Средний R2:  {cv_scores_r2.mean():.4f} (±{cv_scores_r2.std():.4f})")

print("\nРезультаты по фолдам:")
for i in range(5):
    print(f"Фолд {i+1}: MAE = {cv_mae_scores[i]:.4f}, R2 = {cv_scores_r2[i]:.4f}")

# 8. Сохраняем лучшую модель
final_model_rf = best_model
print("\nЛучшая модель сохранена в переменную 'final_model_rf'")
```

#### Обучение модели  XGBRegressor

<div style="background-color: #e6f3ff; color: black; padding: 15px; border-radius: 5px; border-left: 5px solid #4da6ff">
<strong>Комментарий студента</strong>: 

Создаем пайплайн обучения XGB на тренировочных данных, с оптимальным количеством гиперпараметров и оценкой по MAE на кросс-валидации.
    
</div>



```python
# 1. Создаем пайплайн с предобработкой и моделью
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('xgb', XGBRegressor(
        random_state=140725,
        n_jobs=-1,
        objective='reg:squarederror'  # Используем MSE
    ))
])

# 2. Сетка гиперпараметров
param_grid = {
    'xgb__n_estimators': [100, 150],  # Увеличиваем количество деревьев
    'xgb__learning_rate': [0.01, 0.05, 0.1],  # Более тонкий подбор learning rate
    'xgb__max_depth': [3, 5, 7],  # Расширяем диапазон глубины
    'xgb__subsample': [0.7, 0.9],  # Добавляем вариативность субсэмплинга
    'xgb__colsample_bytree': [0.7, 0.9],  # Вариативность для признаков
    'xgb__reg_alpha': [0, 0.1],  # L1 регуляризация
    'xgb__reg_lambda': [0.8, 1]  # L2 регуляризация
}

# 3. Настраиваем кросс-валидацию с MAE и R2
scoring = {
    'mae': make_scorer(mean_absolute_error),
    'r2': make_scorer(r2_score)
}

grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring='neg_mean_absolute_error',  # Выбираем лучшую модель по MAE
    refit=True,
    cv=5,
    n_jobs=-1,
    verbose=0,
    return_train_score=True
)
# 4. Обучаем модель с подбором параметров
print("Начинаем подбор параметров XGBoost...")
grid_search.fit(X_train, y_train)
print("Подбор параметров завершен!")

# 5. Выводим лучшие параметры
print("\nЛучшие параметры модели:")
best_params = grid_search.best_params_
for param, value in best_params.items():
    print(f"{param}: {value}")

# 6. Кросс-валидация на всей тренировочной выборке
print("\nРезультаты кросс-валидации (5 фолдов):")
cv_results = cross_validate(
    grid_search.best_estimator_,
    X_train,
    y_train,
    scoring=scoring,
    cv=5,
    n_jobs=-1
)

# Выводим только средние значения MAE и R2
mean_mae = np.mean(cv_results['test_mae'])
mean_r2 = np.mean(cv_results['test_r2'])

print(f"Средний MAE: {mean_mae:.4f}")
print(f"Средний R²:  {mean_r2:.4f}")

# 7. Оцениваем модель на тренировочных данных
best_model = grid_search.best_estimator_
y_train_pred = best_model.predict(X_train)

metrics = {
    'MAE': mean_absolute_error(y_train, y_train_pred),
    'R2': r2_score(y_train, y_train_pred)
}

print("\nМетрики на полной тренировочной выборке:")
for name, value in metrics.items():
    print(f"{name}: {value:.4f}")

# 8. Сохраняем лучшую модель в переменную
final_model_gb = best_model
print("Лучшая модель также сохранена в переменную 'final_model_gb'")
```
